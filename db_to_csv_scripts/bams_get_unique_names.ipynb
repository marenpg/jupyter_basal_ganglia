```python
import os
import pandas as pd

root_dir = "../Data/csvs/bams2_connectivity/"

def getAllUniqueNames():
    names = set()
    
    for filename in os.listdir(root_dir):
        if filename.endswith(".csv"):
            df_bams_conn = pd.read_csv(root_dir + filename, dtype="object")
            unique_from = df_bams_conn.From.unique()
            unique_to = df_bams_conn.To.unique()
            for name in unique_from:
                names.add(name)
            for name2 in unique_to:
                names.add(name2)

    return sorted(names)

def getBasalGangliaAreas():
    names = set()
    for filename in os.listdir(root_dir):
        if filename.endswith("outputs.csv"):
            df_bams_conn = pd.read_csv(root_dir + filename, dtype="object")
            unique_from = df_bams_conn.From.unique()
            for name in unique_from:
                names.add(name)
    return sorted(names)

sorted_all = getAllUniqueNames()
print("All sorted")
for elem in sorted_all:
    print(elem)
    
sorted_bg = getBasalGangliaAreas()
print()
print("AREAS only Basal ganglia")
for elem in sorted_bg:
    print(elem)



```
